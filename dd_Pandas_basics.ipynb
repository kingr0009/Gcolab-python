```python
#always start with importing libraraies : numpy and Pandas
import numpy as np
import pandas as pd
```

```python
my_series = pd.Series(data = [2,3,5,4], index = ['a','b','c','d'])   #len of data and series should be equal
print(my_series)
```

```python
#you can also create a series from a dictionary, in which case the dictionary acts as the labels
my_dict = {'a':2, 'b':3, 'c':5, 'd':4}
my_series = pd.Series(my_dict)
print(my_series)
```

```python
#we can access items in a series by labels
print(my_series['a'])
```

```python
#numeric index also
print(my_series[0])
```

```python
#if you take slice of a series, you get both the values and the labels contained in the slice
my_series[1:3]
```

```python
#operational
q = my_series*2
w = my_series + my_series

print(q)
print(w)
```

```python
np.mean(my_series)
```

```python
# prompt: write a small code as an example of loc syntax

# Sample data (replace with your actual data)
data = {'name': ['Alice', 'Bob', 'Charlie'],
        'age': [25, 30, 28],
        'salary': [50000, 60000, 55000]}

# Create the DataFrame
df = pd.DataFrame(data)

# Access the row where 'name' is 'Bob' using .loc
bob_row = df.loc[df['name'] == 'Bob']
print(bob_row)

# Access multiple rows based on a condition
rows_above_28 = df.loc[df['age'] > 28]
print(rows_above_28)

# Access a specific cell using .loc
alice_salary = df.loc[df['name'] == 'Alice', 'salary'].iloc[0]
print(alice_salary)
```
