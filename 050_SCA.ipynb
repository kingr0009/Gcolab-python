```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Supplier': ['Supplier A', 'Supplier B', 'Supplier A', 'Supplier C', 'Supplier B', 'Supplier C', 'Supplier A'],
    'Location': ['Location X', 'Location Y', 'Location X', 'Location Z', 'Location Y', 'Location Z', 'Location X'],
    'Transportation Mode': ['Air', 'Sea', 'Air', 'Road', 'Sea', 'Road', 'Air'],
    'SKU': ['SKU1', 'SKU2', 'SKU1', 'SKU3', 'SKU2', 'SKU3', 'SKU1'],
    'Shipping Route': ['Route1', 'Route2', 'Route1', 'Route3', 'Route2', 'Route3', 'Route1'],
    'Revenue': [1000, 1500, 1200, 800, 1600, 900, 1100],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the dataset
print("Dataset:")
print(df)
```

```python
# Count the number of times each supplier is used
supplier_count = df['Supplier'].value_counts().reset_index()
supplier_count.columns = ['Supplier', 'Usage Count']

# Display the supplier usage table
print("\nTable: Suppliers and Usage Count")
print(supplier_count)
```

```python
# Plot a pie chart for supplier usage
plt.figure(figsize=(8, 6))
plt.pie(supplier_count['Usage Count'], labels=supplier_count['Supplier'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green'])
plt.title('Supplier Usage Distribution')
plt.show()
```

```python
# Calculate total revenue generated by each supplier
supplier_revenue = df.groupby('Supplier')['Revenue'].sum().reset_index()

# Display the revenue table
print("\nTable: Revenue Generated by Each Supplier")
print(supplier_revenue)
```

```python
# Plot a pie chart for supplier revenue
plt.figure(figsize=(8, 6))
plt.pie(supplier_revenue['Revenue'], labels=supplier_revenue['Supplier'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green'])
plt.title('Revenue Distribution by Supplier')
plt.show()
```

```python
# Count the number of times each location is used
location_count = df['Location'].value_counts().reset_index()
location_count.columns = ['Location', 'Usage Count']

# Display the location usage table
print("\nTable: Locations and Usage Count")
print(location_count)
```

```python
# Plot a pie chart for location usage
plt.figure(figsize=(8, 6))
plt.pie(location_count['Usage Count'], labels=location_count['Location'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green'])
plt.title('Location Usage Distribution')
plt.show()
```

```python
# Calculate total revenue generated by each location
location_revenue = df.groupby('Location')['Revenue'].sum().reset_index()

# Display the revenue table
print("\nTable: Revenue Generated by Each Location")
print(location_revenue)
```

```python
# Plot a pie chart for location revenue
plt.figure(figsize=(8, 6))
plt.pie(location_revenue['Revenue'], labels=location_revenue['Location'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green'])
plt.title('Revenue Distribution by Location')
plt.show()
```

```python
# Count the number of times each transportation mode is used
transportation_count = df['Transportation Mode'].value_counts().reset_index()
transportation_count.columns = ['Transportation Mode', 'Usage Count']

# Display the transportation usage table
print("\nTable: Transportation Mode and Usage Count")
print(transportation_count)
```

```python
# Calculate total revenue generated by each transportation mode
transportation_revenue = df.groupby('Transportation Mode')['Revenue'].sum().reset_index()

# Display the revenue table for transportation modes
print("\nTable: Revenue Generated by Each Transportation Mode")
print(transportation_revenue)
```

```python
# Count the number of times each SKU is used
sku_count = df['SKU'].value_counts().reset_index()
sku_count.columns = ['SKU', 'Usage Count']

# Display the SKU usage table
print("\nTable: SKU and Usage Count")
print(sku_count)
```

```python
# Calculate total revenue generated by each SKU
sku_revenue = df.groupby('SKU')['Revenue'].sum().reset_index()

# Display the revenue table for SKUs
print("\nTable: Revenue Generated by Each SKU")
print(sku_revenue)
```

```python
# Count the number of times each shipping route is used
route_count = df['Shipping Route'].value_counts().reset_index()
route_count.columns = ['Shipping Route', 'Usage Count']

# Display the shipping route usage table
print("\nTable: Shipping Route and Usage Count")
print(route_count)
```

```python
# Calculate total revenue generated by each shipping route
route_revenue = df.groupby('Shipping Route')['Revenue'].sum().reset_index()

# Display the revenue table for shipping routes
print("\nTable: Revenue Generated by Each Shipping Route")
print(route_revenue)
```

```python
# Plot a pie chart for transportation mode usage
plt.figure(figsize=(8, 6))
plt.pie(transportation_count['Usage Count'], labels=transportation_count['Transportation Mode'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green', 'purple'])
plt.title('Transportation Mode Usage Distribution')
plt.show()
```

```python
# Plot a pie chart for transportation mode revenue
plt.figure(figsize=(8, 6))
plt.pie(transportation_revenue['Revenue'], labels=transportation_revenue['Transportation Mode'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green', 'purple'])
plt.title('Revenue Distribution by Transportation Mode')
plt.show()
```

```python
# Plot a pie chart for SKU usage
plt.figure(figsize=(8, 6))
plt.pie(sku_count['Usage Count'], labels=sku_count['SKU'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green', 'purple'])
plt.title('SKU Usage Distribution')
plt.show()
```

```python
# Plot a pie chart for SKU revenue
plt.figure(figsize=(8, 6))
plt.pie(sku_revenue['Revenue'], labels=sku_revenue['SKU'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'green', 'purple'])
plt.title('Revenue Distribution by SKU')
plt.show()
```
